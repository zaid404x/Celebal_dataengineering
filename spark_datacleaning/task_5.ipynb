```python
df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
df.show(truncate=False)
```

```python
df.printSchema()
```

```python
print("Before:", df.count())

df = df.dropDuplicates()

print("After:", df.count())
```

```python
df = df.fillna({
    "salary": 0,
    "age": 25,
    "region": "Unknown"
})
```

```python
df = df.dropna(subset=["department"])
```

```python
from pyspark.sql.functions import *

df = df.withColumn(
    "name",
    when(col("name") == "", "Unknown")
    .otherwise(col("name"))
)
```

```python
age_filtered = df.filter(
    (col("age") >= 21) &
    (col("age") <= 30)
)

age_filtered.show()
```

```python
north_employees = df.filter(
    col("region") == "North"
)

north_employees.show()
```

```python
df = df.withColumnRenamed(
    "salary",
    "employee_salary"
)
```

```python
df.printSchema()
```

```python
df = df.withColumn(
    "bonus",
    col("employee_salary") * 0.10
)

df.show()
```

```python
df.select(
    count("*").alias("Total Employees")
).show()
```

```python
df.select(
    sum("employee_salary")
).show()
```

```python
df.select(
    avg("employee_salary")
).show()
```

```python
df.select(
    min("employee_salary")
).show()
```

```python
df.select(
    max("employee_salary")
).show()
```

```python
dept_stats = df.groupBy("department").agg(
    count("*").alias("employee_count"),
    sum("employee_salary").alias("total_salary"),
    avg("employee_salary").alias("avg_salary"),
    min("employee_salary").alias("min_salary"),
    max("employee_salary").alias("max_salary")
)

dept_stats.show()
```

```python
dept_stats.filter(
    col("avg_salary") > 55000
).show()
```

```python
dept_stats.orderBy(
    col("avg_salary").desc()
).show()
```

```python
clean_df = (
    df.dropDuplicates()
      .fillna({
          "employee_salary": 0,
          "region": "Unknown",
          "age": 25
      })
)

result = (
    clean_df.groupBy("department")
            .agg(
                avg("employee_salary")
                .alias("avg_salary")
            )
)

result.show()
```
