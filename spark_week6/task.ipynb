```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python

spark = SparkSession.builder \
    .appName("Celebal_Week6") \
    .getOrCreate()

print("Spark Session Created Successfully")
```

```python
df = spark.read.csv(
    "employee.csv",
    header=True,
    inferSchema=True
)
```

```python
df.printSchema()
```

```python
df.show(7)
```

```python
print("Total Records:", df.count())
```

```python
age_df = df.filter(col("age") > 28)
age_df.show()
```

```python
selected_df = df.select(
    "name",
    "age",
    "salary"
)

selected_df.show()
```

```python
df = df.withColumnRenamed(
    "salary",
    "monthly_salary"
)

df.show(5)
```

```python
df = df.withColumnRenamed(
    "salary",
    "monthly_salary"
)

df.show(5
```

```python
df = df.withColumnRenamed(
    "salary",
    "monthly_salary"
)

df.show(11)
```

```python
df = df.withColumn(
    "bonus",
    col("monthly_salary") * 0.10
)

df.show(5)
```

```python
df = df.withColumn(
    "total_compensation",
    col("monthly_salary") + col("bonus")
)

df.show(5)
```

```python
df.select([
    count(
        when(col(c).isNull(), c)
    ).alias(c)
    for c in df.columns
]).show()
```

```python
df = df.na.fill({
    "monthly_salary": 0
})
```

```python
df.select([
    count(
        when(col(c).isNull(), c)
    ).alias(c)
    for c in df.columns
]).show()
```

```python
df.groupBy("department") \
    .avg("monthly_salary") \
  .show()
```

```python
df.orderBy(
    col("monthly_salary").desc()
).show(20)
```

```python
df.orderBy(
    col("monthly_salary").desc()
).show(99)
```

```python
it_df = df.filter(
    col("department") == "IT"
)

it_df.show()
```

```python
df.write.mode("overwrite") \
  .option("header", True) \
  .csv("output_csv")
```

```python
df.write.mode("overwrite") \
  .parquet("output_parquet")
```

```python
parquet_df = spark.read.parquet(
    "output_parquet"
)

parquet_df.show(5)
```

```python
csv_df = spark.read.csv(
    "employee.csv",
    header=True,
    inferSchema=True
)

parquet_df = spark.read.parquet(
    "output_parquet"
)

print(csv_df.count())
print(parquet_df.count())
```
